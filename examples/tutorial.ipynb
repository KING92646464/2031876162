```
import shutil
import os
from google.colab import drive

# 挂载 Google Drive
# 此步骤会要求你登录 Google 账号并获取授权码
# 授权后，Google Drive 会被挂载到 /content/drive 路径下
drive.mount('/content/drive')

# 定义要复制的源文件夹路径
# 请将这里替换为你实际要复制的文件夹路径
source_folder = '/content/runs/detect/train4'

# 定义目标文件夹路径，即 Google Drive 中的保存位置
# 请将这里替换为你希望保存到 Google Drive 的具体路径
destination_folder = '/content/drive/MyDrive/日志1'

# 检查源文件夹是否存在
if os.path.exists(source_folder):
    try:
        # 使用 shutil.copytree 函数复制整个文件夹
        # 此函数会递归地复制源文件夹及其所有子文件夹和文件
        shutil.copytree(source_folder, destination_folder)
        print(f"成功将 {source_folder} 复制到 {destination_folder}")
    except FileExistsError:
        # 如果目标文件夹已经存在，会抛出 FileExistsError 异常
        print(f"目标文件夹 {destination_folder} 已经存在，请删除或选择其他路径。")
    except Exception as e:
        # 捕获其他可能的异常并打印错误信息
        print(f"复制过程中出现错误: {e}")
else:
    # 如果源文件夹不存在，给出相应提示
    print(f"源文件夹 {source_folder} 不存在，请检查路径。")
```
